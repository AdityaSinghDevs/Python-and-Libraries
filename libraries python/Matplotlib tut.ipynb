## BASIC GRAPH

```python
x=[0,2,4,6]
y=[0,1,2,3]

# resize the graph( always at the beginning of plot)
plt.figure(figsize=(2,3), dpi=200)

# plt.plot(x,y, color='r', linewdith=2 , linestyle='--', marker='^', markersize=10, markeredgecolor='b', label='2x',)
# use shorthand
# fmt= '{color][marker][linestyle'
plt.plot(x,y, 'r.--', label= '2x')
plt.legend()

# giving titles and axis labels

plt.title("HERES THE FIRST GRAPH", fontdict={'font':'Comic Sans MS', 'fontsize':'16'})
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')

# controlling ticks
plt.xticks([0,2,4,6,8,10])
plt.yticks([0,2,4,6,8,10])

# using numpy arrays
x2=np.arange(0,4.5,0.5)
# plt.plot(x2, x2**2, 'b^-', label='X^2')

# to show projection add limits

plt.plot(x2[:6], x2[:6]**2, 'b^-', label='X^2')
plt.plot(x2[4:], x2[4:]**2, 'b^--', label='X^2')

# save file
plt.savefig('mygraph.png', dpi=300)

plt.show()
```

## BAR GRAPHS

```python
label=['A','B','C']
values=[2,4,8]
plt.figure(figsize=(5,2))
bars=plt.bar(label,values)

# bar[0].set_hatch('/')
# bar[1].set_hatch('-')
# bar[2].set_hatch('.')

# for cleaner code
patterns=['/','o','*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))
```

## REAL LIFE EXAMPLES

```python
gas = pd.read_csv('gas_prices.csv')
gas

# plt.figure(figsize=(7,5), dpi=100)
# ''lt.plot(gas.Year, gas.USA, '.-', label='USA' )
# plt.plot(gas.Year, gas.Canada, '.-', label='Canada')
# plt.plot(gas.Year, gas.Japan, '.-', label='Japan')
# plt.plot(gas.Year, gas['South Korea'], '.-', label='South Korea')


# Another way to plot MANY VALUES!!!

for country in gas:
    if country!='Year':
        plt.plot( gas.Year, gas[country], marker='.')
    

plt.title('Gas Prices Over the Years', fontdict={'font':'Times New Roman', 'fontsize':20, 'fontweight':'bold'})

plt.legend()

# controlling ticks
plt.xticks(gas.Year[::3])
plt.yticks([0.2,4,6,8])

plt.xlabel('YEARS',fontdict={'font':'Times New Roman', 'fontsize':20, 'fontweight':'bold'})
plt.ylabel('USD',fontdict={'font':'Times New Roman', 'fontsize':20, 'fontweight':'bold'})

plt.savefig('gasgraph.png', dpi=200)

plt

plt.show()
```
